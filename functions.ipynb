Ques 1- What is the difference between a function and a method in python?

Ans 1- In Python, the main difference between a function and a method is that functions are used for independent tasks, while methods are used for actions specific to a class or object.

A method is a function that is defined within a class and is meant to operate on instances of

```python
#Function Example

def greet(name):

 return f"Hello, {name}!"

print(greet("Alice"))
```

Ques 5- What are iterators in python and how do they differ from iterables?

Ans 5- An Iterable is basically an object that any user can iterate over. An Iterator is also an object that helps a user in iterating over another object.

how's they differ given below:

**Iterables**



*  An iterable is any Python object capable of returning its members one at a time, making it usable in a for loop or with other iteration tools like map filter, and list comprehensions.
* Common examples of iterables include data structures like lists, tuples dictionaries, sets, and strings. These objects have an __iter__() method, which allows them to return an iterator when needed.

**Iterators**



* An iterator is a special object that enables Python to fetch items from an iterable one at a time.
*  You can get an iterator from an iterable by calling iter() on it. The iterator itself has a __next__() method,which returns the next item from the sequence each time it’s called. Once the items are exhausted,__next__() raises a StopIteration exception.
* Iterators are stateful, meaning they remember where they are in the iteration process. They’re also "lazy," as they fetch elements only when requested, which is useful for working with large or even infinite sequences.




```python
#Example

my_iterator = iter(my_list)
print(next(my_iterator))
print(next(my_iterator))
```

Ques 6- Explain the concept of generators in python and how they are defined?

Ans 6- A Python generator function is defined like a normal function. However, in the normal function, we use the return keyword, whereas in generators, we use the yield keyword to generate a value. If yield is present in the body of the def, that function will automatically become a generator function.

**Generator Functions**

A generator function is defined like a regular function but includes the yield keyword. Each time yield is
encountered, the function pauses and yields a value to the caller, resuming on the next next() call.


```python
#Example

def countdown(n):
  while n > 0:
    yield n
    n -= 1

for num in countdown(5):
  print(num)
```

**Generator Expressions**

A generator expression is a concise way to create a generator, similar to list comprehensions but with parentheses. It’s useful for quick, on-the-fly generation of data.


```python
#Example

squares = (x ** 2 for x in range(5))
for square in squares:
  print(square)
```

Ques 4- Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
#Ans 4-
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(number, end=' ')
```

Ques 5- Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
#Ans 5-

class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next = 1
        self.index = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.index >= self.terms:
            raise StopIteration
        result = self.current
        self.current, self.next = self.next, self.current + self.next
        self.index += 1
        return result

fib = Fibonacci(10)
for num in fib:
    print(num, end=' ')
```

Ques 6- Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
#Ans 6-

def powers_of_2(limit):
    for i in range(limit + 1):
        yield 2 ** i

for power in powers_of_2(5):
    print(power)
```

```python
#Ans 7-

def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()

for line in read_file_line_by_line('example.txt'):
    print(line)
```

Ques 9- Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

```python
#Ans 9-

def celsius_to_fahrenheit(celsius_list):
    return list(map(lambda c: (c * 9/5) + 32, celsius_list))

celsius = [0, 20, 37, 100]
fahrenheit = celsius_to_fahrenheit(celsius)
print(fahrenheit)
```

Ques 10- Create a Python program that uses filter() to remove all the vowels from a given string.

```python
#Ans 10-

def remove_vowels(string):
    return ''.join(filter(lambda char: char.lower() not in 'aeiou', string))

text = "Hello World"
print(remove_vowels(text))  # Output: "Hll Wrld"
```
